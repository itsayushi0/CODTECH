```python
!pip install reportlab pandas matplotlib
import matplotlib.pyplot as plt
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
```

```python
# Step 4: Change path to your file in Drive
file_path = "/content/drive/MyDrive/Crypto.csv"  # change if stored elsewhere

# Step 5: Load file
df = pd.read_csv(file_path)

```

```python
df['Price.1'] = df['Price.1'].replace(',', '', regex=True).astype(float)

summary = df.describe()
total_rows = len(df)
columns = df.columns.tolist()
max_price = df['Price.1'].max()
min_price = df['Price.1'].min()
avg_price = df['Price.1'].mean()
```

```python
plt.figure(figsize=(8,4))
df['Price.1'].head(20).plot(kind='bar', color='skyblue')
plt.title("Top 20 Prices")
plt.ylabel("Price")
plt.xlabel("Index")
plt.tight_layout()
plt.savefig("chart.png")
plt.close()
```

```python
# Step 8: Create PDF
pdf_file = "/content/drive/MyDrive/Crypto_Report.pdf"
c = canvas.Canvas(pdf_file, pagesize=letter)

c.setFont("Helvetica-Bold", 16)
c.drawString(100, 750, "Crypto Data Analysis Report")

c.setFont("Helvetica", 12)
c.drawString(100, 720, f"Total Rows: {total_rows}")
c.drawString(100, 700, f"Columns: {', '.join(columns)}")
c.drawString(100, 680, f"Max Price: {max_price}")
c.drawString(100, 660, f"Min Price: {min_price}")
c.drawString(100, 640, f"Average Price: {avg_price:.2f}")

c.drawImage("chart.png", 100, 400, width=400, height=200)

c.showPage()
c.save()

print(f"Report saved as {pdf_file}")
```
